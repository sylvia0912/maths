## Special Matrices in Linear Algebra

In Linear Algebra there r special matrices with certain properties that r easy to analyse and to manipulate.

### 1. Singular Matrix
- A singular matrix has a determinant value equal to zero
- The singular matrix does not have an inverse

### 2. Non singular matrix
- A non singular matrix has a determinat whose value is a non zero value.
- A non singular matrix has an inverse matrix.

### 3. Inverse Matrix

- We knew that for a real number, the inverse of the number was the reciprocal of the number, as long as the number wasn't zero.
- There r 2 types of inverse - multiplicative inverse and additive inverse. We refer to multiplicative inverse here.
- In real numbers anything multiplied with 1 is 1. So in matrix world analogous thing is Identity matrix. You multiply any matrix with I, u get same matrix. 
- **The inverse of a square matrix A, denoted by A-1, is the matrix so that the product of A and A-1 is the identity matrix.**

### 4. Orthogonal Matrix

- An orthogonal vector is vectors perpendicular to each other.
- For an orthogonal matrix if u view every row as a vector then all rows are perpendicular/orthogonal to each other. That matrix is orthogonal matrix. So dot product of rows should be 0 since they r orthogonal vectors.

- An orthogonal matrix is NOT symmetric always.

### Orthonormal Matrix
- If rows in a matrix are not just perpendicular to each other but length of each row is also equal to 1 then it is an orthonormal matrix.
- Below 1 * 1 = 1 since A is orthonormal vector, rest all rows 2,3,4  perpendicular to 1 hence dot product is 0. Similarly 2 * 2 = 1 ( as orthonormal vector) rest 2 with 1,3,4 all 0. Hence A.A^T = I

### 5. Symmetric Matrices

- In linear algebra, a symmetric matrix is a square matrix whose transpose is same as matrix itself. Formally, because equal matrices have equal dimensions, only square matrices can be symmetric. The entries of a symmetric matrix are symmetric with respect to the main diagonal. 
- Eigen vectors of a symmetric matrix are orthogonal.

### 6. Positive Semi-Definite Matrix


We study PSD and properties of PSD because usually whatever data we will take we will form a matrix & if it turns out to be PSD it will satisfy all these properties of PSD & help to gain better insights.

- Let A denote an n x n symmetric matrix with real entries and let x denote an n x 1 column vector. Then Q = x'Ax is said to be a quadratic form.

- If Quadratic form of A is non-negative, then A is PSD matrix.



### Properties of Positive Semi-Definite

- In PSD, all eigenvalues are greater than 0.
- In PSD, all eigenvectors are perpendicular to each other.

### 7. Covariance Matrix / Variance covariance matrix.

- A covariance matrix is always a square matrix. Furthermore, it is positive semi-definite, and symmetric.
- Its main diagonal contains variances (i.e., the covariance of each element with itself).
- This matrix is very useful in stochastic modeling and principle component analysis.

**What is variance and covariance**

- Variance is a measure of dispersion and can be defined as the spread of data from the mean of the given dataset. 
- Covariance is calculated between two variables and is used to measure how the two variables vary together.
- Variance covariance matrix is defined as a square matrix where the diagonal elements represent the variance and the off-diagonal elements represent the covariance.
- The covariance between two variables can be positive, negative, and zero. A positive covariance indicates that the two variables have a positive relationship whereas negative covariance shows that they have a negative relationship. If two elements do not vary together then they will display a zero covariance.


### 8. Diagnol Matrix

- In linear algebra, a diagonal matrix is a matrix in which the entries outside the main diagonal are all zero; the term usually refers to square matrices. 
- Elements of the main diagonal can either be zero or nonzero
- Diagonal matrices are typically, but not always, square.

### 9. Identity Matrix

- An identity matrix is a square matrix having 1s on the main diagonal, and 0s everywhere else.
