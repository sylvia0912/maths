##  Optimization
### Mathematical programming is referred to as  optimization.
- Mathematical optimization or **mathematical programming**.
- Optimization problems, as the name implies, deal with finding the best, or “optimal” (hence the name) solution to some type of problem, generally mathematical.
- In order to find the optimal solution, we need some way of measuring the quality of any solution. This is done via what is known as an objective function, with “objective” used in the sense of a goal.
- Optimization is a mathematical technique for finding a maximum or minimum value of a function of several variables subject to a set of constraints using calculus (derivatives).


- Optimization is the process of making a quantity as large or small as possible. 
- Optimization problems will always ask you to maximize or minimize some quantity like Find the largest, find the minimum, what dimensions will give the greatest?
- Optimization methods are used in many areas of study to find solutions that maximize or minimize some study parameters, such as minimize costs in the production of a good or service, maximize profits, minimize raw material in the development of a good, or maximize production.

## Goal of Optimization
### Goal made out of the 3 components of optimization which is decision variables, objective function, constraints.
The basic goal of the optimization process is to find decision variable values that minimize or maximize the objective function (result in best objective function) while satisfying the constraints. This result is called an optimal solution.


### 3 Components of mathematical Optimization

Reference - Linear Optimization with Applications http://home.ubalt.edu/ntsbarsh/opre640a/partviii.htm


### **Objective Function**: 

- The objective function in a mathematical optimization problem is the real-valued function whose value is to be either minimized or maximized over the set of feasible alternatives.
- The function of the decision variables that is to be optimised is called the objective function.


### **Decision Variables**: 

- The controllable inputs are the set of decision variables which affect the value of the objective function. In the manufacturing problem, the variables might include the allocation of different available resources, or the labor spent on each activity. 
- **Decision variables are essential. If there are no variables, we cannot define the objective function and the problem constraints.**
- The uncontrollable inputs are called parameters.


### **Constraints**:

- A set of constraints allows some of the decision variables to take on certain values, and exclude others. For the manufacturing problem, it does not make sense to spend a negative amount of time on any activity, so we constrain all the "time" variables to be non-negative.

Optimization models have three major components: 
1. controllable decision variables/design variables
2. objective function
3. constraints

### Deterministic vs Stochastic Optimization


1. Deterministic global optimization is a branch of numerical optimization which focuses on finding the global solutions of an optimization problem whilst providing theoretical guarantees that the reported solution is indeed the global one, within some predefined tolerance.
2. Stochastic optimization is the process of maximizing or minimizing the value of a mathematical or statistical function when one or more of the input parameters is subject to randomness. The word stochastic means involving chance or probability.

### Convex and Non-Convex Optimization

1. Convex optimization there can be only one optimal solution, which is globally optimal or you might prove that there is no feasible solution to the problem
2. Nonconvex optimization may have multiple locally optimal points and it can take a lot of time to identify whether the problem has no solution or if the solution is global.
3. A convex problem usually is much more easier to deal with in comparison to a non convex problem which takes a lot of time and it might lead you to a dead end.
